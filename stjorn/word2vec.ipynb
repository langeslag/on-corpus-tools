# Word2Vec: _Stjórn_ Word Embeddings

The word embeddings model `Word2Vec` takes stock of "what company words keep" and consequently allows us to calculate the "meaning" of words relative to all other words in the corpus. This notebook runs it on the totality of _Stjórn_ itself, to the exclusion of any control corpus. Cf. `doc2vec.ipynb` for a study of document similarity on the same principles.

```python
import os,glob
from gensim.models import Word2Vec
```

```python
def normalize(target):
    # This dict limits orthographical variation beyond the rule sets
    # of stjorn-extract.ipynb and menota-extract.ipynb:
    matrix = {
        'j': 'i',
        'v': 'u',
        'ð': 'þ',
        'á': 'a',
        'ǽ': 'æ',
        'é': 'e',
        'í': 'i',
        'ó': 'o',
        'ú': 'u',
        'ý': 'y',
        'ǿ': 'ø',
        'k': 'c', # rather than vice versa, because of Latin (e.g. Lucifer)
        '[': '',
        ']': ''
        }
    for k,v in matrix.items():
        target = target.replace(k, v)
    return target
```

`Word2Vec` takes sentences as input. _Stjórn_ as extracted from heimskringla.no is divided into paragraphs, not sentences, and a sentence tokenizer is unlikely to do well on our unpunctuated data, so we'll feed the model paragraphs instead.

```python
corpus = []
for i in glob.glob('nlp/*txt'):
    document = open(i).read().splitlines()
    for paragraph in document:
        corpus.append(normalize(paragraph))
paragraphs = [[token for token in document.split()] for document in corpus]
```

```python
model = Word2Vec(sentences=paragraphs, min_count=1, vector_size=300, workers=2, window=3, epochs=30)
```

```python
model.wv.most_similar('dauid')
```

The model has correctly determined that the terms most meaningfully similar to "dauid" are other names, Saul first of all. This may serve as our proof of method.
